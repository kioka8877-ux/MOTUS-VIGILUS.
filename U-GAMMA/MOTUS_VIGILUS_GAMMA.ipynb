# ⚔️ MOTUS-VIGILUS — Frégate U-GAMMA : La Forge
> Manifestation : Données .npz → Animation .fbx (Roblox R15)

**Pipeline** : 📦 .npz → [Blender Headless] → [Rigging R15] → [Bake Animation] → 🎮 .fbx

**Instructions** :
1. Exécuter l'installation (une seule fois)
2. Uploader vos fichiers .npz (produits par la Frégate U-ALPHA)
3. Lancer la forge
4. Télécharger les fichiers .fbx
5. Importer les .fbx dans Roblox Studio

```python
# ═══════ INSTALLATION — FRÉGATE U-GAMMA ═══════
!pip install -q numpy
!sudo apt-get install -y blender > /dev/null 2>&1

import os
if not os.path.exists("MOTUS-VIGILUS"):
    !git clone https://github.com/kioka8877-ux/MOTUS-VIGILUS..git MOTUS-VIGILUS

os.chdir("MOTUS-VIGILUS/U-GAMMA")
os.makedirs("inputs", exist_ok=True)
os.makedirs("outputs", exist_ok=True)
print("✅ Frégate U-GAMMA — Installation terminée")
```

```python
# ═══════ UPLOAD .NPZ ═══════
from google.colab import files
import shutil

uploaded = files.upload()
npz_files = []
for name in uploaded.keys():
    dest = f"inputs/{name}"
    shutil.move(name, dest)
    npz_files.append(dest)
    print(f"✅ {name} → {dest}")

print(f"\n📦 {len(npz_files)} fichier(s) .npz uploadés")
```

```python
# ═══════ LA FORGE — FRÉGATE U-GAMMA ═══════
import glob

npz_files = sorted(glob.glob("inputs/motus_core_P*.npz"))
if not npz_files:
    print("❌ Aucun fichier .npz trouvé dans inputs/")
    print("   Uploadez d'abord les .npz produits par la Frégate U-ALPHA")
else:
    fbx_files = []
    for npz in npz_files:
        fbx_name = os.path.basename(npz).replace(".npz", ".fbx").replace("motus_core_", "MOTUS_VIGILUS_")
        fbx_out = f"outputs/{fbx_name}"
        print(f"\n🔥 Forge : {npz} → {fbx_out}")
        !blender -b templates/r15_template.blend -P codebase/motus_forge.py -- "{npz}" "{fbx_out}"
        fbx_files.append(fbx_out)

    print(f"\n✅ {len(fbx_files)} fichier(s) .fbx forgés")
    for f in fbx_files:
        print(f"  🎮 {f}")
```

```python
# ═══════ TÉLÉCHARGEMENT .FBX ═══════
from google.colab import files
import glob

fbx_files = sorted(glob.glob("outputs/MOTUS_VIGILUS_P*.fbx"))
for f in fbx_files:
    files.download(f)
    print(f"⬇️ {f}")

print("\n⚔️ Frégate U-GAMMA — Forge terminée.")
print("📋 Importer les .fbx dans Roblox Studio → Avatar Animation Editor")
```
