```python
# ═══════ UPLOAD VIDÉO ═══════
from google.colab import files
import shutil

uploaded = files.upload()
video_name = list(uploaded.keys())[0]
video_path = f"inputs/{video_name}"
shutil.move(video_name, video_path)
print(f"✅ Vidéo uploadée : {video_path}")
```

```python
# ═══════ CONFIGURATION ═══════
# @title Paramètres MOTUS-VIGILUS
FPS_CIBLE = 30  # @param [30, 60, 120] {type:"integer"}
LISSAGE = "moyen"  # @param ["faible", "moyen", "brutal"] {type:"string"}
ROOT_MOTION = True  # @param {type:"boolean"}

print(f"⚙️ FPS: {FPS_CIBLE} | Lissage: {LISSAGE} | Root Motion: {ROOT_MOTION}")
```

```python
# ═══════ FRÉGATE U-ALPHA : EXTRACTION ═══════
root_flag = "" if ROOT_MOTION else "--no-root-motion"
!python core/motus_extract.py "{video_path}" -o outputs/ --fps {FPS_CIBLE} --smooth {LISSAGE} {root_flag}

import glob
npz_files = sorted(glob.glob("outputs/motus_core_P*.npz"))
print(f"\n✅ {len(npz_files)} fichier(s) .npz extraits")
for f in npz_files:
    print(f"  📦 {f}")
```

```python
# ═══════ FRÉGATE U-GAMMA : LA FORGE ═══════
import glob

npz_files = sorted(glob.glob("outputs/motus_core_P*.npz"))
fbx_files = []

for npz in npz_files:
    fbx_out = npz.replace(".npz", ".fbx").replace("motus_core_", "MOTUS_VIGILUS_")
    print(f"\n🔥 Forge : {npz} → {fbx_out}")
    !blender -b templates/r15_template.blend -P core/motus_forge.py -- "{npz}" "{fbx_out}"
    fbx_files.append(fbx_out)

print(f"\n✅ {len(fbx_files)} fichier(s) .fbx forgés")
for f in fbx_files:
    print(f"  🎮 {f}")
```

```python
# ═══════ TÉLÉCHARGEMENT ═══════
from google.colab import files
import glob

fbx_files = sorted(glob.glob("outputs/MOTUS_VIGILUS_P*.fbx"))
for f in fbx_files:
    files.download(f)
    print(f"⬇️ {f}")

print("\n⚔️ MOTUS-VIGILUS — Mission accomplie.")
```
